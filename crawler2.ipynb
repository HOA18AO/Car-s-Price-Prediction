```python
headers = [
    'title',
    'date',
    'location',
    'year',
    'used',
    'kms',
    'import', # nhap khau
    'style',
    'gearBox',
    'fuelEngine',
    'exterior',
    'interior',
    'seats',
    'doors',
    'motivated',
    'url'
]
df = pandas.DataFrame(columns=headers)
df
```

```python
df = pandas.read_csv('bonbanh_v3_1111.csv')
df
```

```python
webUrl = 'https://bonbanh.com/'
```

```python
carLinks_df = pandas.read_csv('car_link.csv')
carLinks_df
```

```python
carLinks_df.loc[0:8]
```

```python
carLinks = carLinks_df.values.tolist() # dataframe to 2d list
carLinks = [item for sublist in carLinks for item in sublist] # 2d list to 1d list
carLinks
```

```python
len(carLinks)
```

```python
df_tmp = pandas.DataFrame(columns=headers)
df_tmp
```

```python

for link in carLinks[13000:15000]:
    data = crawlOneCar.crawler(url=webUrl+link)
    newRow = pandas.DataFrame(data, index=[0])
    df_tmp = pandas.concat([df_tmp, newRow], ignore_index=True)
    sleep(2)
```

```python
df_tmp
```

```python
df = pandas.concat([df, df_tmp], ignore_index=True)
```

```python
df.info()
```

```python
df.to_csv('bonbanh_v3_1111.csv', index=False)
```
